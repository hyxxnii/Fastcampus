```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## 계층적 군집분석 

```python
X = np.array([[5,3],
    [10,15],
    [15,12],
    [24,10],
    [30,30],
    [85,70],
    [71,80],
    [60,78],
    [70,55],
    [80,91],])
```

```python
import matplotlib.pyplot as plt

labels = range(1, 11)
plt.figure(figsize=(10, 7))
plt.subplots_adjust(bottom=0.1)
plt.scatter(X[:,0],X[:,1], label='True Position')

for label, x, y in zip(labels, X[:, 0], X[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-3, 3),
        textcoords='offset points', ha='right', va='bottom')
plt.show()

# [1,2,3,4,5] 와 [6,7,8,9,10]으로 2개 군집 나눌 거라고 예상
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt

linked = linkage(X, 'single')
```

```python
labelList = range(1, 11)

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='top',
            labels=labelList,
            distance_sort='descending',
            show_leaf_counts=True)
plt.show()

# 위 예상과 똑같이 [1,2,3,4,5] 와 [6,7,8,9,10] 2개의 군집으로 나뉘었지
# 그 다음에는 [1,2,3,4,5]에서 [1,2,3,4]가 [5]보다 더 가까이 분포해있기때문에
# 그 다음 분포는 [1,2,3,4]와 [5]가 묶인 것을 알 수 있지
# [6,7,8,9,10]에서도 마찬가지로 [6,7,8,10]끼리 분포해있고 [9]는 좀 떨어져있으니까
# 그 다음 분포는 [6,7,8,10]와 [9]
```

```python
# 쇼핑 데이터로 계층적 군집분석 해보자
customer_data = pd.read_csv('./data/shopping-data.csv')
```

```python
customer_data.head()
```

```python
data = customer_data.iloc[:, 3:5].values
```

```python
import scipy.cluster.hierarchy as shc

plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))

# 아랫부분은 잘 볼 수 없다는게 단점
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)

# cluster를 5개로 짤라서 각각 데이터가 어느 cluster에 할당되었는가 확인
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')

# 데이터의 첫번째와 두번째 칼람을 가지고와서 각각의 cluster label별로 색깔 넣어서 확인
```

## boston house 집값 데이터


```python
# 데이터 불러오기
data = pd.read_csv("./data/Boston_house.csv") 

'''
타겟 데이터
1978 보스턴 주택 가격
506개 타운의 주택 가격 중앙값 (단위 1,000 달러)

특징 데이터
CRIM: 범죄율
INDUS: 비소매상업지역 면적 비율
NOX: 일산화질소 농도
RM: 주택당 방 수
LSTAT: 인구 중 하위 계층 비율
B: 인구 중 흑인 비율
PTRATIO: 학생/교사 비율
ZN: 25,000 평방피트를 초과 거주지역 비율
CHAS: 찰스강의 경계에 위치한 경우는 1, 아니면 0
AGE: 1940년 이전에 건축된 주택의 비율
RAD: 방사형 고속도로까지의 거리
DIS: 직업센터의 거리
TAX: 재산세율'''

data.head() # 데이터 확인
```

```python
target=data['Target']
data = data.drop(['Target'], axis = 1) 
```

```python
data.shape
```

```python
plt.figure(figsize=(10, 7))
plt.title("Customer Dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))

# 일단 크게 2개의 군집으로 나뉘었기 때문에 2개로 나뉘어서 확인해보자
```

```python
from sklearn.cluster import AgglomerativeClustering

cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')
cluster.fit_predict(data)
```

```python
np.mean([x for x, y in zip(target, cluster.fit_predict(data)) if  y == 0])
# 0으로 할당된 target값들의 평균값
```

```python
np.mean([x for x, y in zip(target, cluster.fit_predict(data)) if  y == 1])
# 1으로 할당된 target값들의 평균값

# y값 안쓰고 x만 가지고 clustering 했는데도
# 2개의 cluster에 해당되는 y값들의 차이가 분명하게 나는 것을 볼 수 있음
```

```python
data.iloc[cluster.fit_predict(data)==0, :].describe()
# 0번째로 할당된 cluster에는 집값이 낮은 데이터들이 할당된 것을 알 수 있음
```

```python
data.iloc[cluster.fit_predict(data)==1,:].describe()
# 1번째로 할당된 cluster에는 집값이 높은 데이터들이 할당된 것을 알 수 있음

# 각각 0번째랑 1번째로 할당된 데이터들의 통계치를 뽑으면 그 데이터의 특성을 어느정도 파악 가능
# 주택땅(B), 방수(RM)는 큰 차이가 없지만 범죄율(CRIM)에서 큰 차이가 있지
# 0번 cluster는 범죄율이 1번 cluster보다 훨씬 높지 , 다른 변수들도 마찬가지로 확인 가능

# 그렇지만 clustering은 분류 목적보다는 군집 나눠서 군집의 관계를 규명하거나 군집의 특성을 규명하는 데에 쓰인다!! 
```
