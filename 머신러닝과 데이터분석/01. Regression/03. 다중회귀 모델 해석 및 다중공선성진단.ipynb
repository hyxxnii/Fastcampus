```python
import os
import pandas as pd 
import numpy as np
import statsmodels.api as sm
```

```python
# 데이터 불러오기
boston = pd.read_csv("./csv/Boston_house.csv")
boston.head()
```

```python
boston_data = boston.drop(['Target'],axis=1)
# boston_data
```

```python
'''
타겟 데이터
1978 보스턴 주택 가격
506개 타운의 주택 가격 중앙값 (단위 1,000 달러)

특징 데이터
CRIM: 범죄율
INDUS: 비소매상업지역 면적 비율
NOX: 일산화질소 농도
RM: 주택당 방 수
LSTAT: 인구 중 하위 계층 비율
B: 인구 중 흑인 비율
PTRATIO: 학생/교사 비율
ZN: 25,000 평방피트를 초과 거주지역 비율
CHAS: 찰스강의 경계에 위치한 경우는 1, 아니면 0
AGE: 1940년 이전에 건축된 주택의 비율
RAD: 방사형 고속도로까지의 거리
DIS: 직업센터의 거리
TAX: 재산세율'''
```

# crim, rm, lstat을 통한 다중 선형 회귀분석

```python
x_data=boston[['CRIM','RM','LSTAT']] ##변수 여러개
target = boston[['Target']]
x_data.head()
```

```python
x_data1 = sm.add_constant(x_data, has_constant='add')
```

```python
multi_model = sm.OLS(target,x_data1)
fitted_multi_model=multi_model.fit()
```

```python
fitted_multi_model.summary()
```

# crim, rm, lstat, b, tax, age, zn, nox, indus 변수를 통한 다중선형회귀분석

```python
## bostan data에서 원하는 변수만 뽑아오기 
x_data2 = boston[['CRIM', 'RM', 'LSTAT' ,'B', 'TAX', 'AGE', 'ZN', 'NOX', 'INDUS']]
x_data2.head()
```

```python
# 상수항 추가
x_data2_= sm.add_constant(x_data2, has_constant="add")
```

```python
# 회귀모델 적합
multi_model2 = sm.OLS(target, x_data2_)
fitted_multi_model2 = multi_model2.fit()
```

```python
# 결과 출력
fitted_multi_model2.summary()
```

```python
# 세 변수만 추가한 모델의 회귀 계수 
fitted_multi_model.params
```

```python
# full모델의 회귀계수
fitted_multi_model2.params
```

- 6개의 변수가 추가 됐음에도 불구하고 3개의 변수를 가지고 만든 모델의 r-squared에 비해 r-squared 값이 굉장히 적게 올라갔음
- 둘 중에 하나를 골라야한다면 당연히 3개의 변수를 가지고 만든 모델을 선택(parameter가 적은 게 훨씬 좋은 모델임)
- 6개 변수의 모델을 살펴보면 3개 변수 모델과 달리 'CRIM'의 p-valued가 높아졌음 
- 'NOX', 'INDUS'는 p-valued가 높기때문에 변수를 제거해도됨, 제거했을 때 나머지 p-valued는 당연히 낮아짐 ('NOX', 'INDUS'가 가져간 y의 변동성을 다시 다른 변수들이 가져갈 수 있기때문에(변수들간의 변동성 중복은 안된다고했지) 전체적으로 p-valued가 낮아지게되어있음)
- 'AGE'나 'CRIM' 같은 경우는 애매한 경우라 분석자의 주관을 개입시켜서 변수를 넣을지 뺄지 결정
- 그 지역의 범죄율이 집값이 영향을 끼칠 확률이 일반적으로 높으니까 변수를 넣는게 낫다고 판단할수있지

```python
# base모델과 full모델의 잔차비교 
import matplotlib.pyplot as plt
fitted_multi_model.resid.plot(label="full")
fitted_multi_model2.resid.plot(label="full_add")
plt.legend()

'''
잔차가 그렇게 많이 줄어들지않았음
단순선형에서 다중선형으로 넘어갔을 때 잔차가 많이 낮아진 것을 볼 수 있었는데
지금 보면 그렇게 큰 차이는 없더라
'''
```

# VIF를 통한 다중공선성 확인 

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(x_data2.values, i) for i in range(x_data2.shape[1])]
vif["features"] = x_data2.columns
```

```python
vif

'''
'- NOX'가 VIF가 굉장히 크게 나온 것을 알 수 있음
- 이 변수를 y로 보고 나머지 변수들로 회귀분석을 했을 때
  나머지 변수들로 이 변수를 설명하는 비율(변동의 비율)을 r-squared로 계산
- VIF가 10이상이면 다중공선성이 있다고 진단
- 굉장히 많은 변수들이 다중공선성이 있음
- 'NOX', 'INDUS' 변수를 제거하는 게 맞고 다시 분석을 하는 식으로
'''
```

```python
## nox 변수 제거후(X_data3) VIF 확인 

vif = pd.DataFrame()
x_data3= x_data2.drop('NOX',axis=1)
vif["VIF Factor"] = [variance_inflation_factor(
    x_data3.values, i) for i in range(x_data3.shape[1])]
vif["features"] = x_data3.columns
vif

'''
제거 후 다른 변수들의 VIF는 조금이라도 줄어듦
'NOX'를 제거했음에도 불구하고 다중공선성을 가진 변수가 많지
'RM'가 다른 변수들보다 VIF가 좀 크지
산점도를 확인해보면 'RM'은 'LSTAT'와 correlaton이 좀 강해보임 
VIF만 보면 지워야하는 것은 많지만 이전에 봤던 정보들까지 보면 여기서 멈추는게 맞음

일단 'RM'을 제거해보자
'''
```

```python
## RM 변수 제거후(x_data4) VIF 확인 

vif = pd.DataFrame()
x_data4= x_data3.drop('RM',axis=1)
vif["VIF Factor"] = [variance_inflation_factor(x_data4.values, i) for i in range(x_data4.shape[1])]
vif["features"] = x_data4.columns
vif

'''
'RM'을 지웠더니 아직도 다중공선성이 있는 변수들이 있지만 이전에 비해서는 굉장히 감소했지
이정도 쯤에 멈추고 다시 회귀분석 진행해보자
'''
```

```python
# nox 변수 제거한 데이터(x_data3) 상수항 추가 후 회귀 모델 적합
# nox, rm 변수 제거한 데이터(x_data4) 상수항 추가 후 회귀 모델 적합

x_data3_ = sm.add_constant(x_data3, has_constant="add")
x_data4_ = sm.add_constant(x_data4, has_constant="add")

model_vif = sm.OLS(target, x_data3_)
fitted_model_vif = model_vif.fit()
model_vif2 = sm.OLS(target, x_data4_)
fitted_model_vif2 = model_vif2.fit()
```

```python
## 회귀모델 결과 비교 
fitted_model_vif.summary()

'''
'NOX'변수 제거한 모델의 r-squared 값은 이전 (3개 변수, 9개 변수 적용) 모델과 크게 차이없지
'''
```

```python
fitted_model_vif2.summary()

'''
3개의 변수만 넣었을 때와 9개 변수 넣었을 때의 r-squared값이 거의 차이가 안났는데
지금 변수를 1개('RM') 제거했는데 r-squared가 굉장히 많이 감소했지 
그럼 그 변수는 굉장히 중요한 변수, 즉 제거하면 안되는 변수라는 것
'RM'변수는 제거하지 말았어야 할 변수라는 것
회귀모델 성능지표인 AIC만 봐도 'RM'변수 제거 안한 모델이 더 낮지 (낮을 수록 좋음)
따라서 'RM'을 제거 안한 위의 모델이 더 좋은 모델이라고 판단 가능
'''
```

# 학습 / 검증데이터 분할

```python
from sklearn.model_selection import train_test_split
X = x_data2_ # 9개의 변수
y = target
train_x, test_x, train_y, test_y = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state = 1)
print(train_x.shape, test_x.shape, train_y.shape, test_y.shape)
```

```python
# train_x 회귀모델 적합
train_x.head()
fit_1 = sm.OLS(train_y, train_x)
fit_1 = fit_1.fit() # 덮어씌기할 때 주의할 점은 재실행하면 또 덮어씐다는 것!
```

```python
## 검증데이터 에대한 예측값과 true값 비교 
plt.plot(np.array(fit_1.predict(test_x)),label="pred")
plt.plot(np.array(test_y),label="true")
plt.legend()
plt.show()

# 잘 된건지 안된건지 모르겠단말이지,,?
```

```python
## x_data3와 x_data4 학습 검증데이터 분할
# 1개의 변수 제거한 거과 2개의 변수 제거한 것에 대해서 학습 검증데이터 분할
X = x_data3_
y = target
train_x2, test_x2, train_y2, test_y2 = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 1)
```

```python
X = x_data4_
y = target
train_x3, test_x3, train_y3, test_y3 = train_test_split(X, y, train_size = 0.7, test_size = 0.3, random_state = 1)
```

```python
test_y.head()
```

```python
test_y2.head()

# test_y1과 test_y2의 index가 같지 (random_state 값을 동일하게 줬기때문에, 비교를 위해서)
```

```python
# x_data3/x_data4 회귀 모델 적합(fit2,fit3)
fit_2 = sm.OLS(train_y2, train_x2)
fit_2 = fit_2.fit()

fit_3 = sm.OLS(train_y3, train_x3)
fit_3 = fit_3.fit()
```

```python
## true값과 예측값 비교 
plt.plot(np.array(fit_2.predict(test_x2)),label="pred2") # 변수 1개 제거
plt.plot(np.array(fit_3.predict(test_x3)),label="pred3") # 변수 2개 제거
plt.plot(np.array(test_y2),label="true")
plt.legend()
plt.show()

# 변수 1개 제거한 것이 변수 2개 제거한 것보다 더 나은 걸 보이고있다라는 걸 알 수 있음
```

```python
## full모델 추가해서 비교 
plt.plot(np.array(fit_1.predict(test_x)),label="pred")
plt.plot(np.array(fit_2.predict(test_x2)),label="pred_vif")
plt.plot(np.array(fit_2.predict(test_x2)),label="pred_vif2")
plt.plot(np.array(test_y2),label="true")
plt.legend()
plt.show()

# 2개 변수 제거한 게 성능이 젤 안 좋은 걸 알 수 있지 (pred_vif2)
```

```python
plt.plot(np.array(test_y2['Target']-fit_1.predict(test_x)),label="pred_full") # 실제값과 예측값의 차이니까 잔차
plt.plot(np.array(test_y2['Target']-fit_2.predict(test_x2)),label="pred_vif")
plt.plot(np.array(test_y2['Target']-fit_3.predict(test_x3)),label="pred_vif2")
plt.legend()
plt.show()

# 2개 변수 제거한 게 잔차가 젤 큰 것을 알 수 있지 (pred_vif2)
```
